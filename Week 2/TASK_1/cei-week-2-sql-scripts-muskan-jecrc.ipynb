```python
import duckdb

df = pd.read_csv(
    '/kaggle/input/datasets/vivek468/superstore-dataset-final/Sample - Superstore.csv',
    encoding='latin1'
)
```

```python
duckdb.register('superstore', df)
```

```python
# Q.1
duckdb.sql("""
SELECT *
FROM superstore
""").df()
```

```python
# Q.2
duckdb.sql("""
SELECT DISTINCT "Customer Name", City
FROM superstore
""").df()
```

```python
# Q.3
duckdb.sql("""
SELECT DISTINCT Category 
FROM superstore;
""").df()
```

```python
#Q.7
duckdb.sql("""SELECT "Order ID", "Customer Name", "Ship Mode" 
FROM superstore 
WHERE "Ship Mode" = 'Standard Class' 
""").df()
```

```python
#Q.8
duckdb.sql(""" SELECT "Product ID", "Product Name", Sales 
FROM superstore 
WHERE Category = 'Technology' 
  AND Sales > 2000 
LIMIT 5;
""").df()
```

```python
#Q.9
duckdb.sql("""
SELECT DISTINCT "Customer Name", State, "Order Date" 
FROM superstore 
WHERE State = 'Maharashtra' 
  AND "Order Date" BETWEEN '2024-01-01' AND '2024-12-31' 
LIMIT 5;
""").df()
```

```python
#Q.10
duckdb.sql("""SELECT "Order ID", "Order Date", "Customer Name", Sales 
FROM superstore 
WHERE "Order Date" BETWEEN '2024-08-10' AND '2024-08-25' 

""").df()
```

```python
#Q.12
duckdb.sql("""SELECT * FROM superstore
WHERE "Order Date" >= '2024-01-01' AND "Order Date" <= '2024-12-31';
""").df()
```

```python
#Q.13
duckdb.sql("""SELECT COUNT(DISTINCT "Order ID") AS Total_Unique_Orders 
FROM superstore;
""").df()
```

```python
#Q.14
duckdb.sql ("""SELECT SUM(Sales) AS Total_Revenue 
FROM superstore;
""").df()
```

```python
#Q.15
duckdb.sql ("""SELECT Category, AVG(Sales / Quantity) AS Avg_Unit_Price 
FROM superstore 
GROUP BY Category;
""").df()
```

```python
#Q.16
duckdb.sql ("""SELECT "Ship Mode", 
       COUNT(DISTINCT "Order ID") AS Order_Count, 
       SUM(Sales) AS Total_Revenue 
FROM superstore 
GROUP BY "Ship Mode" 
ORDER BY Total_Revenue DESC;
""").df()
```

```python
#Q.17
duckdb.sql("""SELECT Category, 
       MAX(Sales / Quantity) AS Most_Expensive, 
       MIN(Sales / Quantity) AS Cheapest 
FROM superstore 
GROUP BY Category;
""").df()
```

```python
#Q.18
duckdb.sql("""SELECT Category, AVG(Sales) AS Avg_Sales 
FROM superstore 
GROUP BY Category 
HAVING AVG(Sales) > 200;
""").df()
```

```python
#Q.24
duckdb.sql("""SELECT "Product Name", 
       Sales AS Unit_Price, 
       CASE 
            WHEN Sales < 50 THEN 'Budget' 
            WHEN Sales BETWEEN 50 AND 300 THEN 'Mid-Range' 
            ELSE 'Premium' 
       END AS price_tier 
FROM superstore ;
""").df()
```

```python
#Q.25
duckdb.sql("""
SELECT COUNT(CASE WHEN "Ship Mode" = 'First Class' THEN 1 END) AS First_Class_Count, 
       COUNT(CASE WHEN "Ship Mode" <> 'First Class' THEN 1 END) AS Other_Modes_Count 
FROM superstore;
""").df()
```
